```python
                                                    #........import every library........#

import cv2
#cv2: OpenCV library used for image processing.

import numpy as np
#numpy: Used for numerical operations (like arrays).

import matplotlib.pyplot as plt
#matplotlib.pyplot: Used to display images in a notebook.





                                                    #........Load image........#

image = cv2.imread('my_unicorn.jpg')
#cv2.imread(...): Loads the image from file (your unicorn image).

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#cv2.cvtColor(...) is used to convert it to RGB so it displays correctly in matplotlib.
#OpenCV loads images in BGR color format by default.





                                                   #........Show original image........#

plt.imshow(image)
#Displays the original image using matplotlib.

plt.title("Original Image")
plt.axis('off')
#.axis('off') hides the X-Y axis.

plt.show()
#.show() renders the image in Jupyter.





                                                    #........User Input........#

action = input("\nWhat do you want to do?\nOptions: black and white / rotate / flip / sketch\n ")
#Asks user what operation they want to perform.





                                                    #........Black & White........#

if action.lower() == "black and white":
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
#Converts the RGB image into grayscale (black and white).
                                               
    plt.imshow(gray_image, cmap='gray')
    plt.title("Black and White Image")
    plt.axis('off')
    plt.show()
#Shows the black & white image using a gray colormap.





                                                    #........Rotate Image........#

elif action.lower() == "rotate":
    angle = int(input("How much do you want to rotate? (e.g., 45/90/180/... Degree): "))
#Asks the user how many degrees to rotate the image.
    
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
#Gets the height (h) and width (w) of the image.
#Sets the center point of the image (needed for rotation).
    
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
#Creates a rotation matrix for the specified angle.
    
    rotated = cv2.warpAffine(image, M, (w, h))
#Rotates the image using the matrix created above.
    
    plt.imshow(rotated)
    plt.title(f"Rotated Image by {angle}°")
    plt.axis('off')
    plt.show()
#Displays the rotated image.





                                            #........Flip Image........#

elif action.lower() == "flip":
    direction = input("In which direction? (horizontally/vertically): ").lower()
#Asks if user wants to flip horizontally or vertically.

    if direction == "horizontally":
#Horizontally
        
        flipped = cv2.flip(image, 1)
#1 flips the image left-right.

        plt.imshow(flipped)
        plt.title("Horizontally Flipped Image")
        plt.axis('off')
        plt.show()
#Displays the Horizontally-flipped image.
    
    elif direction == "vertically":
#Vertically 
        
        flipped = cv2.flip(image, 0)
#0 flips the image up-down.
        
        plt.imshow(flipped)
        plt.title("Vertically Flipped Image")
        plt.axis('off')
        plt.show()
#Displays the Vertically-flipped image.
    
    else:
        print("Invalid flip direction!")
#If type another Direction.





                                                #........Sketch Effect........#

elif action.lower() == "sketch":
#Make sketch
    
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
#Convert to grayscale.
    
    inv = 255 - gray
#Inverts color: dark becomes light, light becomes dark.
    
    blur = cv2.GaussianBlur(inv, (21, 21), 0)
#Apply blur to smooth the image (creates the pencil effect).
    
    sketch = cv2.divide(gray, 255 - blur, scale=256)
#Final sketch effect using blending.

    plt.imshow(sketch, cmap='gray')
    plt.title("Sketch Image")
    plt.axis('off')
    plt.show()
#Show the sketch.





                                                #........If Nothing........#
else:
    print("Invalid option! Please Choose Write Option...")
#If the input is not one of the options, show an error.
```
